## Theano

```python
!pip install Theano
```

```python
import theano.tensor as T
from theano import function
```

```python
# Declaring 2 variables
x = T.dscalar('x')
y = T.dscalar('y')
```

```python
# Subtracting the 2 numbers
z = x - y
```

```python
# Converting it to a callable object so that it takes matrix as parameters
f = function([x, y], z)
```

```python
f(10, 7)
```
